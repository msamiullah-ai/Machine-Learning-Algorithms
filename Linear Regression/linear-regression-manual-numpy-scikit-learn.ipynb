```python
import numpy as np
from sklearn.linear_model import LinearRegression
import math
```

### Univariate Linear Regression using Loops

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost = 0
    for i in range(m):
        f_wb = w*x[i] + b
        error = y - f_wb
        cost+= (error**2)

    actual_cost = cost/(2*m)
    return actual_cost
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    d_dw = 0
    d_db = 0
    for i in range(m):
        f_wb = w*x[i] + b
        error = f_wb - y[i]
        d_dw += error * x[i]
        d_db += error
    dj_dw = d_dw/m
    dj_db = d_db/m
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, alpha, num_iter, costfunc, gradientfunc):
    J_history = []
    p_history = []
    w = w_in
    b = b_in

    for i in range(num_iter):
        dj_dw, dj_db = gradientfunc(x, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
    
    if i < 100000:
        J_history.append(costfunc(x, y, w, b))
        p_history.append([w, b])

    if i % math.ceil(num_iter / 10) == 0:
        print(f"Iteration {i:5}: "
                f"Cost {J_history[-1]:0.2e} "
                f"dj_dw: {dj_dw:0.3e} "
                f"dj_db: {dj_db:0.3e} "
                f"w: {w:0.3e} "
                f"b: {b:0.3e}")
    return w, b, J_history, p_history
```

### Multivariate Linear Regression using Loops

Pick one row (i).

            For that row:

            multiply feature 1 by w[0]

            multiply feature 2 by w[1]

            multiply feature 3 by w[2]

            Add them all together.

            Add the bias b.

            That's the prediction for that row

            Compare the prediction with y[i] to compute the error.

Move to the next row and repeat.

```python
def compute_cost(X, y, w, b):
    m = X.shape[0]
    n = X.shape[1]

    cost = 0

    for i in range(m):

        f_wb = 0

        for j in range(n):
            f_wb += w[j] * X[i][j]

        f_wb += b

        error = f_wb - y[i]
        cost += error ** 2

    return cost / (2 * m)
```

```python
def compute_gradient(X, y, w, b):

    m, n = X.shape

    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):

        # Compute prediction for row i
        f_wb = 0
        for j in range(n):
            f_wb += w[j] * X[i][j]

        f_wb += b

        # Error for this training example
        err = f_wb - y[i]

        # Gradient for each weight
        for j in range(n):
            dj_dw[j] += err * X[i][j]

        # Gradient for bias
        dj_db += err

    # Average gradients
    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_db, dj_dw
```

```python
def gradient_descent(X, y, w_in, b_in, alpha, num_iter, costfunc, gradientfunc):

    J_history = []
    p_history = []

    w = w_in.copy()
    b = b_in

    for i in range(num_iter):

        # Compute gradients
        dj_db, dj_dw = gradientfunc(X, y, w, b)

        # Update parameters
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        # Save cost
        if i < 100000:
            J_history.append(costfunc(X, y, w, b))
            p_history.append([w.copy(), b])

        # Print progress every 10%
        if i % math.ceil(num_iter / 10) == 0:
            print(f"Iteration {i:5}: "
                  f"Cost {J_history[-1]:0.2e} "
                  f"dj_dw: {dj_dw} "
                  f"dj_db: {dj_db:0.3e} "
                  f"w: {w} "
                  f"b: {b:0.3e}")

    return w, b, J_history, p_history
```

### Vectorized Version Using Numpy

```python
def compute_cost(X, y, w, b):
    m = X.shape[0]

    predictions = X @ w + b
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)

    return cost
```

```python
def compute_gradient(X, y, w, b):
    m = X.shape[0]

    predictions = X @ w + b
    error = predictions - y

    dj_dw = (1 / m) * (X.T @ error)
    dj_db = (1 / m) * np.sum(error)

    return dj_db, dj_dw
```

```python
def gradient_descent(X, y, w, b, alpha, num_iters):

    J_history = []

    for i in range(num_iters):

        dj_db, dj_dw = compute_gradient(X, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        J_history.append(compute_cost(X, y, w, b))

    return w, b, J_history
```

### Linear Regression Using Scikit Learn

```python
# Training data
X = np.array([
    [2104, 5, 45],
    [1416, 3, 40],
    [1534, 3, 30],
    [852,  2, 36]
])

y = np.array([460, 232, 315, 178])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Learned parameters
print("Weights (w):", model.coef_)
print("Bias (b):", model.intercept_)

# Predict
new_house = np.array([[1800, 3, 20]])
prediction = model.predict(new_house)

print("Prediction:", prediction)
```
